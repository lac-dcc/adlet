```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("sparse_sparse_old.csv")
len(data)
```

```python
data = data[data["mean_exec_time"] >= 0]
len(data)
```

## (0.5, 0.5) * (0.5, 0.5)

```python
df = data
df = df[(df["left_row_sparsity"] == 0.5) & (df["left_col_sparsity"] == 0.5)& (df["right_row_sparsity"] == 0.5) & (df["right_col_sparsity"] == 0.5)]
best_time = df.nsmallest(5, 'mean_exec_time')
worst_time = df.nlargest(5, 'mean_exec_time')
#print(best_time)
#print(worst_time)
top_and_bottom = pd.concat([best_time, worst_time])
top_and_bottom['left_right_format'] = top_and_bottom['left_format'] + ' - ' + top_and_bottom['right_format']
top_and_bottom.plot.bar(x=('left_right_format'), y='mean_exec_time', title='(0.5, 0.5) * (0.5, 0.5)')
plt.show()
data
```
